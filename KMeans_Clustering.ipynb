```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('/content/diabetes.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe().loc['mean'].plot.bar()
```

```python
plt.figure(figsize=(10,8))
sb.heatmap(df.corr(),annot=True)
```

```python
plt.scatter(df['Glucose'].values,df['Age'].values)
```

```python
km=KMeans(n_clusters=2)
km.fit(df)
```

```python
pred=km.predict(df)
pred
```

```python
plt.scatter(df['Glucose'].values,df['Insulin'].values,c=pred)
```

```python
sb.scatterplot(x="Glucose",y="Age",hue="Outcome",data=df,palette="viridis")
```
